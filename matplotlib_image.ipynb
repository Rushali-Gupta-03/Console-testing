```python

import matplotlib.pyplot as plt
from warpdrive import WarpDrive


wd = WarpDrive()
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create the plot
plt.plot(x, y, label='Prime numbers', color='blue', marker='o')

# Add titles and labels
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.grid(True)
wd.save_image(plt)
```
